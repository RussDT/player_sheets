```python
import pandas as pd

game_id=21300019
year = 2014
def four_factors(game_id,year):

    season =str(year-1)+'-'+str(year)[-2:]
    df =pd.read_csv(str(year)+'/'+str(game_id)+'.csv')
    df.drop_duplicates(subset='PLAYER_ID',inplace=True)
    avg = pd.read_csv('https://raw.githubusercontent.com/gabriel1200/site_Data/refs/heads/master/avg_shooting.csv')
    df['Season']=season
    avg=avg[['ORB%','ORtg','Season']]
    
    avg.columns=['avg_orb%','avg_ortg','Season']
    
    #df= df.merge(avg,on='Season')
    df['FG2A']=df['FGA']-df['FG3A']
    df['FG2M']= df['FGM']-df['FG3M']
    df['FG2_miss']=df['FG2A']-df['FG2M']
    
    df['FG3_miss']=df['FG3A']-df['FG3M']
    df['FG2_points']=df['FG2M']*2
    df['FG3_points']=df['FG3M']*3
    sumframe=df.groupby(['TEAM_ID','TEAM_ABBREVIATION']).sum(numeric_only=True)[['POINTS','FGA','FGM','FG2A','FG2M','FG3A','FG3M','OREB','DREB','TOV','FTA','FTM','FG2_miss','FG3_miss','FG3_points','FG2_points']].reset_index()
    print(sumframe['OREB'])
    sumcopy=sumframe[['TEAM_ID','DREB','TOV']].reset_index()
    
    opp_oreb=sumcopy['DREB'].tolist()
    opp_tov=sumcopy['TOV'].tolist()
    
    sumcopy['opp_DREB']=opp_oreb[::-1]
    sumcopy['opp_TOV']=opp_tov[::-1]
    sumcopy.drop(columns=['DREB','TOV'],inplace=True)
    sumframe['Season']=season
    df= sumframe.merge(avg,on='Season')
    df=df.merge(sumcopy,on=['TEAM_ID'])
    df['avg_ortg']/=100
    df['avg_orb%']/=100
    df['2shooting_factor']=df['FG2_points']-(df['FG2M']*df['avg_ortg']) - (1-df['avg_orb%'])*df['FG2_miss']*df['avg_ortg']
    
    df['3shooting_factor']=df['FG3_points']-(df['FG3M']*df['avg_ortg']) - (1-df['avg_orb%'])*df['FG3_miss']*df['avg_ortg']
    
    df['turnover_factor']=-(df['TOV']*df['avg_ortg'])
    df['oreb_factor']=((1-df['avg_orb%'])*df['OREB']-df['avg_orb%']*df['opp_DREB'])*df['avg_ortg']
    df['ft_factor']=df['FTM']-.4*df['FTA']*df['avg_ortg'] + .06*(df['FTA']-df['FTM'])*df['avg_ortg']
    df['spread']=df['ft_factor']+df['oreb_factor']+df['turnover_factor']+df['2shooting_factor']+df['3shooting_factor']
    return df
```
